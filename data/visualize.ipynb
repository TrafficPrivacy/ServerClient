```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
distances = []
time = []
with open("server-astar-5000.csv", newline="") as csvfile:
    reader = csv.reader(csvfile, delimiter=",")
    for row in reader:
        try:
            t = float(row[5]) / 1e9
            if t < 100:
                distances.append(float(row[4]))
                time.append(t)
        except ValueError:
            continue
```

```python
plt.title("A star 5000 runs")
plt.scatter(distances, time)
plt.ylabel("time, (s)")
plt.xlabel("distance")
plt.show()
```

```python
overlaps = []
with open("client1.csv", newline="") as csvfile:
    reader = csv.reader(csvfile, delimiter=",")
    for row in reader:
        try:
            overlap = int(row[4])
            overlaps.append(overlap)
        except ValueError:
            continue
```

```python
plt.title("path overlapping")
plt.hist(overlaps)
plt.xlabel("number of overlaps")
plt.ylabel("number of segments")
plt.show()
```
