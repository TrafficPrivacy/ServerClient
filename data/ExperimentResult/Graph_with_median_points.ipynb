```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
def read(path):
    f=open(path,'r')
    return f.readlines()
```

```python
def analyze(filename,rate):
    lines=read(filename)
    rate=float(rate)
    dic={}
    length=len(lines)
    for i in range(1,length):
        line=lines[i]
        line=line.split()
        x=int(line[2])
        y=int(line[1])
        if x not in dic:
            dic[x]=[]
            dic[x].append(y)
        else:
            dic[x].append(y)
    for x in dic:
        dic[x].sort()
    newx=[]
    newy=[]
    for x in dic:
        newx.append(x)
        size=len(dic[x])
        newy.append(dic[x][int(size*rate)])
    tmp=np.poly1d(np.polyfit(newx,newy,2))
    fitted_x=np.linspace(0,max(newx),100)
    fitted_y=tmp(fitted_x)
    plt.scatter(newx,newy,s=5)
    x2=[0,max(newx)]
    y2=[0,x2[1]]
    plt.plot(x2,y2,color='green')
    plt.plot(fitted_x,fitted_y,'-',color='red')
    plt.show()
```

```python
analyze("count_version=1_threshold=0.9.txt",0.5)
```

```python
analyze("count_version=2_threshold=0.9.txt",0.5)
```

```python
analyze("count_version=3_threshold=0.9.txt",0.5)
```
