```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def read(path):
    f=open(path,'r')
    return f.readlines()
```

```python
def main(argv):
    lines=read(argv)
    x=[]
    y=[]
    length=len(lines)
    for i in range(1,length):
        line=lines[i]
        line=line.split()
        if(int(line[2])==1):
            continue
        x.append(int(line[2]))
        y.append(int(line[1]))
    tmp=np.poly1d(np.polyfit(x,y,3))
    fitted_x=np.linspace(0,max(x),100)
    fitted_y=tmp(fitted_x)
    plt.scatter(x, y,s=4)
    x2=[0,max(x)]
    y2=[0,x2[1]]
    plt.plot(x2,y2,color='green')
    plt.plot(fitted_x,fitted_y,'-',color='red')
    plt.show()
```

```python
main("count_version=1_threshold=0.9.txt")
```
